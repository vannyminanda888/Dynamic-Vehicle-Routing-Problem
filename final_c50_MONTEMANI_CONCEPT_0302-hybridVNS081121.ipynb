```python
def timing(f):
    @wraps(f)
    def wrap(*args, **kw):
        ts = time()
        result = f(*args, **kw)
        te = time()
        print ('func:took: %2.4f sec' % \
          (te-ts))
        return result
    return wrap
```

```python
# random.seed(2)
# random.seed(2)
max_vehicle=50
vehicle_capacity=160
depot_close_time=351
time_step=0
service_time=15
class dvrp:
    def __init__(self,data,number_of_runs,number_of_iterations,HMCR,PAR,HMS):
        self.nodes=[x for x in range (1,len(data))]
        self.demand=[x[1] for x in data ]
        self.x=[x[2] for x in data ]
        self.y=[x[3] for x in data ]
        self.section=[x[4] for x in data ]
        #self.service_time=[x[6] for x in data]
        self.time_avail_sec=[x[5] for x in data ]#appearance time
        self.vehicle_capacity=160
        self.depot_close_time=351
        self.Time_budget=351
        self.n_ts=25
        self.t_ac=0.1*self.depot_close_time
        self.service_time_col=[x[6] for x in data ]
        self.service_time=15
        self.nodes_num=50

        # FOR HSA
        self.number_of_runs = number_of_runs
        self.number_of_iterations = number_of_iterations
        self.HMCR = HMCR
        self.PAR = PAR
        self.HMS = HMS
        
        self.historic_improvement_operators={
            'swap_within':[],
            'swap_between':[],
            'insertion_between':[],
            'shuffle_within':[],
        }
        
    def distance(self,a,b):
        distance=math.sqrt(((self.x[a]-self.x[b])**2)+(self.y[a]-self.y[b])**2)
        return distance
    
    
    def calculate_distance_list(self,test_list):  
        res = list(zip(test_list, test_list[1:])) 
        total=0
        for i in range (len(test_list)-1):
            total+=self.distance(res[i][0],res[i][1])
        return total
    

    def swap_operator (self, list_, limit):
        lista=copy.deepcopy(list_)
        try:
            rand1, rand2 = random.sample(range(limit+1, len(lista)-1), 2)
        
        # Can't take two nodes none or 1 are not commited yet.
        except:
            return lista
        lista[rand1],lista[rand2]=lista[rand2],lista[rand1]
        return lista

    
    #shuffle the order within soln

    def rearrange_list(self, list_, limit):
        results=[]
        
        check_list_length = list_[limit+1:]
        
        if len(check_list_length) > 3: 
        
            while True:
                positions = random.sample([x for x in range (limit+1,len(list_)-1)], 2)
                min_position, max_position = min(positions), max(positions)
                if max_position-min_position>1:
                    break

            segment = list_[min_position:max_position]

            for i in range (len(segment)):
                segment_to_shuffle = copy.deepcopy(segment)
#                 random.shuffle(segment_to_shuffle)
                result = list_[0:min_position]+segment_to_shuffle+list_[max_position:]
                results.append(result)

            return results
        
        else:
            
            return [list_]
        
    def shuffle_operator (self, lista, limit):
        try:
            list_ = copy.deepcopy(lista)
            trytry=data.rearrange_list(list_,limit)
            best_soln = []
            best_fitness = 999999
            for i in trytry:
    #             print(i)
                dist = data.calculate_distance_list(i)
    #             print(dist)
                if dist < best_fitness:
                    best_fitness = dist
                    best_soln = copy.deepcopy(i)
        except:
            return list_
        
        return best_soln
        
    def insertion_between(self, route1, route2, limit1, limit2):
        
        route1_, route2_ = copy.deepcopy(route1), copy.deepcopy(route2)
        
        try:
            val_swap1, val_swap2 = random.choice(route1_[limit1+1:-1]), random.choice(route2_[limit2+1:-1])
            
        # Everything is commited so choosing two nodes is impossible
        except:
            return route1_, route2_
        
        idx_route1 = route1_.index(val_swap1)
        idx_route2 = route2_.index(val_swap2)
        
        nodes_before = route1_ + route2_
        repeated_nodes_before = [item for item, count in Counter(nodes_before).items() if count > 1]
        

        #move node from route 1 to route 2 position
        route1_ = route1_[0:idx_route1]+route1_[idx_route1+1:]
        route2_ = route2_[0:idx_route2+1]+[val_swap1]+route2_[idx_route2+1:]
        
        nodes_after = route1_ + route2_
        repeated_nodes_after = [item for item, count in Counter(nodes_after).items() if count > 1]
        
        if len(repeated_nodes_after) > len(repeated_nodes_before):
            print("\nPROBLEM IS IN INSERTION WOHOOOOOOOOOOOOOOOOOOOOOOOOOOOO")
            print('nodes before were: ', nodes_before)
            print('nodes after are: ', nodes_after)
        
        return route1_, route2_
    
    
    def swap_between(self, route1, route2, limit1, limit2):
        
        route1_, route2_ = copy.deepcopy(route1), copy.deepcopy(route2)
        
        try:
            val_swap1, val_swap2 = random.choice(route1_[limit1+1:-1]), random.choice(route2_[limit2+1:-1])
            
            
        # Everything is commited so choosing two nodes is impossible
        except:
            return route1_, route2_
        
        idx_route1 = route1_.index(val_swap1)
        idx_route2 = route2_.index(val_swap2)
        
        nodes_before = route1_ + route2_
        repeated_nodes_before = [item for item, count in Counter(nodes_before).items() if count > 1]
        

        #move node from route 1 to route 2 position
        route1_[idx_route1], route2_[idx_route2] = route2_[idx_route2],route1_[idx_route1]
               
        nodes_after = route1_ + route2_
        repeated_nodes_after = [item for item, count in Counter(nodes_after).items() if count > 1]
        
        if len(repeated_nodes_after) > len(repeated_nodes_before):
            print("\nPROBLEM IS IN swap between")
            print('nodes before were: ', nodes_before)
            print('nodes after are: ', nodes_after)
        
        return route1_, route2_    
    
    def route_shuffle(self):
        self.nodes.remove(0)
#         random.shuffle(self.nodes)
        route_shuffle=self.nodes[:]
        return route_shuffle
    
    
    def capacity_sum(self,list_):
        current_cap=0
        for i in (list_):
            current_cap+=self.demand[i]
        return current_cap
    
    
    def calculate_distance(self):
        self.dictio={}
        a=self.route_shuffle()
        for i in (a):
            dtc=self.distance(0,i)
            self.dictio[i]=dtc
        return self.dictio
    
    
    ##########this function calculate the total time of a list(include service time and travel distance)#########
    def calculate_current_time(self,list_):
        service_time=0
        distance=0
        total_time=0
        lista=list_[:]
        distance=dvrp.calculate_distance_list(self,lista)
        for i in (list_):
            service_time+=self.service_time_col[i]
        total_time=service_time+distance
        return total_time

    
    def index_empty_route(dict):
        for i, (k,v) in enumerate (dict.items()):
            if len(v)<=2:
                return k

            
    def remove_empty_route(dict,key):
        del dict[key]
        return dict

    
    def find_closest_node(self):
        ll=self.calculate_distance()
        self.closest=min(ll.keys(), key=(lambda k: ll[k]))
        return self.closest
   

    def find_completed_node(self, time_step, route, new_comer):
        current_visited_route=[]
        for i in route:
            current_visited_route.append(i)
            current_time=data.calculate_current_time(current_visited_route)
            if current_time<time_step:
                completed_route=copy.deepcopy(current_visited_route)
            limit=int(len(completed_route))#the length let say 3 but the position is position 2 so minus 1 here
        return limit
       
        
    def calculate_fitness_from_dictionary(self,dictionary):
        fitness=0
        for i in range (len(dictionary)):
            list_=dictionary[i+1]
            time=dvrp.calculate_current_time(self, list_)
            fitness+=time
        return fitness
    
    
    def sort_nodes_based_appearance(self):
        self.appearance_dict={}
        for i in range (len(self.nodes)):
            self.appearance_dict['{}'.format(self.nodes[i])]=self.time_avail_sec[i]
        return self.appearance_dict #return the sorted node based on its appearance time
    
    
    def calc_current_timeslice_commitment_time(self):
        T=self.depot_close_time
        commitment_time=self.current_time+(T/self.n_ts)+self.t_ac
        return commitment_time
          
    
    def calculate_next_time_slice(self):
        ''' This function is used to update the current time slice'''
        T=self.depot_close_time
        self.current_time+=(T/self.n_ts)
        
        
    
    def add_static_nodes_back_to_soln(self):
        ''' Appends static nodes into solutions dictionary before applying HSA'''
        
        for soln in self.dictionary_harmony_memory.keys():         
            
            # Shuffle since we will always pop the last node
#             random.shuffle(self.static_customers_dict[soln])
            
            vehicle = 1
            
            # While you still have static customers
            while True:
                
                # If no more customers on current soln, move to next soln
                if len(self.static_customers_dict[soln]) == 0:
                    break
                
                # Check if vehicle exist; if it does, make a copy to avoid overwriting
                if str(vehicle) in self.dictionary_harmony_memory[soln].keys():   
                    # Append node
                    temporary_route = copy.deepcopy(self.dictionary_harmony_memory[soln][str(vehicle)])
                 
                #if vehicle does not exist (means that we need to assign additional vehicle to served route:)
                else:
                    temporary_route = [0,0]
                    
                # Obtain customer to serve and append to temporary route
                customer = self.static_customers_dict[soln].pop()
                temporary_route.insert(-1, customer)

                #Check if capacity constraint is violated
                capacity = self.capacity_sum(temporary_route)
                
                # If violated, put customer back to dictionary and increase vehicle number
                if capacity > self.vehicle_capacity:
                    self.static_customers_dict[soln].append(customer)
                    vehicle+=1
                
                # If first constraint is not violated, check second constraint
                else:
                    
                    # #Check if time constraint is violated
                    current_traveling_time=self.calculate_current_time(temporary_route)
                    
                    # If violated, put customer back to dictionary and increase vehicle number
                    if current_traveling_time>self.depot_close_time:
                        self.static_customers_dict[soln].append(customer)
                        vehicle+=1

                    #if both capacity & travel time limit not exceeded then save temporary route into dict
                    else:
                        
                        self.dictionary_harmony_memory[soln][str(vehicle)]=temporary_route
                        
                        capacity = self.capacity_sum(temporary_route)
                        current_traveling_time=self.calculate_current_time(temporary_route)
#                         print(f"this is the route {self.dictionary_harmony_memory[soln]} with travel_dist {current_traveling_time} and cap {capacity}\n\n")
                    

    def update_static_customers(self):
        ''' This function recalculates the static customers based on the non-commited + new dynamic'''
        
        self.static_customers_dict = {}
        self.commited_nodes_dict = {}
        self.remaining_nodes={}
        
        for soln in self.dictionary_harmony_memory.keys():
            
            # Check for repeated nodes!
            lista=data.dictionary_harmony_memory[soln]
            combine=[]
            for vehicle,route in lista.items():
                combine+=route
                
            a = [item for item, count in Counter(combine).items() if count > 1]
            
            if (len(a) > 1):
                print("SOLN: ", data.dictionary_harmony_memory[soln])
                print("AHA!")
            

            self.commited_nodes_dict[soln]={}
            temp_remains_list=[]
            
            # Find Commited Nodes
            for route_idx in self.dictionary_harmony_memory[soln].keys():
                route = self.dictionary_harmony_memory[soln][route_idx]
                idx = self.find_commited_nodes(route)
                
                # If the index is a number (something has been commited)
                if idx:
                    self.commited_nodes_dict[soln][route_idx]=idx
                
                    # Update Harmony Memory Solution (remove non-commited nodes)
                    try:
                        temp_remains_list+=route[idx+1:-1]
                        self.dictionary_harmony_memory[soln][route_idx] = route[0:idx+1]+[0]
                        self.remaining_nodes[soln]=temp_remains_list

                    except:
                        self.dictionary_harmony_memory[soln][route_idx] = route
                     
                # Nothing is commited inside the route    
                else:
                    temp_remains_list+=[x for x in route if x!=0]
                    self.dictionary_harmony_memory[soln][route_idx] = [0,0]
                    self.remaining_nodes[soln]=temp_remains_list
                    self.commited_nodes_dict[soln][route_idx]=0
                    

            # Update Static Nodes
            scheduled = [y for x in self.dictionary_harmony_memory[soln].values() for y in x if y]
            #static = [x for x in self.nodes if self.appearance_dict[str(x)] > self.depot_close_time*0.5 and x not in scheduled]
            self.static_customers_dict[soln] = self.remaining_nodes[soln]+[x for x in self.nodes if self.appearance_dict[str(x)]<self.current_time and x not in scheduled and x not in self.remaining_nodes[soln]]

            
    def find_commited_nodes(self,route):
        '''
        This iterates over all nodes in a vehicle to find which nodes are commited.
        '''
        distance=0
        commitment_time_threshold = self.calc_current_timeslice_commitment_time()
        
        for idx in range(1,len(route)-1):
            pair=route[idx-1:idx+1]
            distance+=self.distance(pair[0],pair[1])+self.service_time
            
            if distance>commitment_time_threshold:
                return idx
        
        # If we come from previous time_slices it means we must have some commited nodes
        if distance > ((self.depot_close_time/self.n_ts)+self.t_ac):
            return int(len(route)-2)
        
        # We are on the first time slice
        else:
            return None
        
        
    def initial_solution(self):
        
        ''' Initial Variables '''
        self.current_time=0
        vehicle=0

        initial_route_capacity=0
        tco = 0.5 #cut off time here only
        cut_off=self.Time_budget*tco
        dict_=data.sort_nodes_based_appearance()
        
        open_nodes=[]
        self.commited_dictionary = {}
#         dictionary_route={}
        self.dictionary_harmony_memory={}
        dictionary_static_nodes={}
        
        #Population Generation
        for solution in range (self.HMS):
            dictionary_route={}
            static_nodes=[]
            vehicle=0
            # Step 2&3: Create StaticProb
            for key, v in dict_.items():
                if v > cut_off:
                    keys=int(key)
                    static_nodes.append(keys)
#             random.shuffle(static_nodes) # shuffle

        
            # Step 4: Set Vehicle to Depot
            initial_route=[0,0]#depot


            # Step 5: Assign nodes to routes

            #5.1: Assign static nodes to routes
            for node in static_nodes:
                temporary_route=copy.deepcopy(initial_route)
                
                #here we check which vehicle has the lowest cost to append a new node
                
                temporary_route.insert(-1,int(node))
                temporary_route_capacity=data.capacity_sum(temporary_route)
                temporary_traveling_time=data.calculate_current_time(temporary_route)
                if initial_route_capacity<=vehicle_capacity: 
                    if temporary_traveling_time<=depot_close_time:
                        initial_route=copy.deepcopy(temporary_route)
                    else:
                        vehicle+=1
                        #we update the solution to the dictionary
                        dictionary_route['{}'.format(vehicle)]=initial_route 
                        #now we append the other solution into other vehicle
                        initial_route=[0,0]
                        initial_route.insert(-1,int(node))

                else:
                    vehicle+=1
                    dictionary_route['{}'.format(vehicle)]=initial_route 
                    #now we append the other solution into other vehicle
                    initial_route=[0,0]
                    initial_route.insert(-1,int(node))

            #this is for the last iteration
            vehicle+=1
            
            nodes_already_in_dict = [y for x in dictionary_route.values() for y in x]
            nodes_already_in_dict = [x for x in nodes_already_in_dict if x!=0]
            temporary_route = [x for x in temporary_route if x not in nodes_already_in_dict ]
            dictionary_route['{}'.format(vehicle)]=temporary_route 
#             print(dictionary_route)#it return the initial solution dictionary
            self.dictionary_harmony_memory['solution {}'.format(solution)]=dictionary_route


    
    
    def generate_random_solution(self):
        
        ''' Initial Variables '''
        self.current_time=0
        vehicle=0

        initial_route_capacity=0
        tco = 0.5 #cut off time here only
        cut_off=self.Time_budget*tco
        dict_=data.sort_nodes_based_appearance()
        
        open_nodes=[]
        self.commited_dictionary = {}
#         dictionary_route={}
        self.dictionary_harmony_memory={}
        dictionary_static_nodes={}
        #Population Generation
        for solution in range (1):
            dictionary_route={}
            static_nodes=[]
            vehicle=0
            # Step 2&3: Create StaticProb
            for key, v in dict_.items():
                if v > cut_off:
                    keys=int(key)
                    static_nodes.append(keys)
#             random.shuffle(static_nodes) # shuffle

        
            # Step 4: Set Vehicle to Depot
            initial_route=[0,0]#depot


            # Step 5: Assign nodes to routes

            #5.1: Assign static nodes to routes
            for node in static_nodes:
                temporary_route=copy.deepcopy(initial_route)
                temporary_route.insert(-1,int(node))
                temporary_route_capacity=data.capacity_sum(temporary_route)
                temporary_traveling_time=data.calculate_current_time(temporary_route)
#                 if initial_route_capacity<=vehicle_capacity: 
                if temporary_route_capacity<=vehicle_capacity: 
                    if temporary_traveling_time<=depot_close_time:
                        initial_route=copy.deepcopy(temporary_route)
                    else:
                        vehicle+=1
                        #we update the solution to the dictionary
                        dictionary_route['{}'.format(vehicle)]=initial_route 
                        #now we append the other solution into other vehicle
                        initial_route=[0,0]
                        initial_route.insert(-1,int(node))

                else:
                    vehicle+=1
                    dictionary_route['{}'.format(vehicle)]=initial_route 
                    #now we append the other solution into other vehicle
                    initial_route=[0,0]
                    initial_route.insert(-1,int(node))

            #this is for the last iteration
            vehicle+=1
            
            nodes_already_in_dict = [y for x in dictionary_route.values() for y in x]
            nodes_already_in_dict = [x for x in nodes_already_in_dict if x!=0]
            temporary_route = [x for x in temporary_route if x not in nodes_already_in_dict ]
            dictionary_route['{}'.format(vehicle)]=temporary_route 
            return dictionary_route
    
    
    ''' 
                                                
                ############################### HSA FUNCTIONS ################################ 
                                                
    '''
    
    def create_fitness_dict(self):
        ''' Creates the fitness dictionary before applying HSA for each time slice'''
        self.fitness_dict = {}
        for soln in self.dictionary_harmony_memory.keys():
            obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
            self.fitness_dict[soln]=obj
    
    
    def find_worst_soln(self):
        ''' Finds the worst soln in HM (based only on TT)'''
        minimum_key =  max(self.fitness_dict, key=self.fitness_dict.get)
        return minimum_key, self.fitness_dict[minimum_key]
    
    def two_opt(self, list1,limit1):
#         print('entered two_opt')
    
        # Obtain segment of nodes that are not fixed in their positions
        temp_route = copy.deepcopy(list1)
        segment = temp_route[limit1+1:-1]

        if len(segment) >= 2:
            # Grab 2 indeces
            a,b = random.sample(segment,2)
            a_idx, b_idx = temp_route.index(a), temp_route.index(b)

            # Make sure b_idx > a_idx
            a_idx, b_idx = min(a_idx,b_idx), max(a_idx,b_idx)

            # Get result
            result = temp_route[:a_idx]+temp_route[a_idx:b_idx+1][::-1]+temp_route[b_idx+1:]
#             print('left two opt 1')
            return result

        else:
#             print('left two opt 2')
            return list1
           
    
    def optimize(self):
        
        #vns parameters:
        self.neighbor_number = 10
        self.counter_vns = 0
        vns_improvement = []
        ''' Performs HSA Optimization'''
        counter_swap_within = 0
        counter_swap_between = 0
        counter_shuffle = 0
        counter_insertion = 0
        counter_not_enter_par = 0
        
        
        self.vns_2opt_counter = 0
        self.vns_insertion_counter = 0
        for run in range(self.number_of_runs):

            for iteration in range(self.number_of_iterations):
                rand1=random.uniform(0,1)
                soln = random.choice(list(self.dictionary_harmony_memory.keys()))                
                # HMCR
                if rand1<self.HMCR:
#                     soln = random.choice(list(self.dictionary_harmony_memory.keys()))
                    rand2=random.uniform(0,1)
                    
                    # PAR
                    if rand2<self.PAR:
                        rand3=random.uniform(0,1)
                        
                        # Apply Swap Within Route
                        if rand3 <= 0.3:
                            
                            # Choose a Vehicle
                            route_idx = random.choice(list(self.dictionary_harmony_memory[soln]))
                            
                            route = self.dictionary_harmony_memory[soln][route_idx]
                            
#                             # SWAP
#                             if len(route) < 0.5:
                            route_obj_before_swap = self.calculate_current_time(route)

                            # Swap
                            try:
                                route_swapped = self.swap_operator(route, self.commited_nodes_dict[soln][route_idx])

                            except KeyError:
                                route_swapped = self.swap_operator(route, 0)

                            route_obj_after_swap = self.calculate_current_time(route_swapped)
                            route_cap_after_swap=self.capacity_sum(route_swapped)
                            # Update if you improved
                            if route_obj_after_swap<self.depot_close_time:
                                if route_cap_after_swap<self.vehicle_capacity:
                                    if route_obj_after_swap < route_obj_before_swap:
                                        self.dictionary_harmony_memory[soln][route_idx]=route_swapped
                                        obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
                                        self.fitness_dict[soln] = obj
                                        counter_swap_within+=((route_obj_before_swap-route_obj_after_swap)/route_obj_before_swap)
                                        
                        # re-arrange list(shuffle and consider all of the combination)
#                         elif 0.3<=rand3<0.6:
#                             route_idx = random.choice(list(self.dictionary_harmony_memory[soln]))
                            
#                             route = self.dictionary_harmony_memory[soln][route_idx]
#                             if len(route)>5:
#                                 route_obj_before_shuffle = self.calculate_current_time(route)

#                                 # sshuffle
#                                 try:
#                                     route_shuffled = self.shuffle_operator(route, self.commited_nodes_dict[soln][route_idx])

#                                 except KeyError:
#                                     route_shuffled = self.shuffle_operator(route, 0)

#                                 route_obj_after_shuffle = self.calculate_current_time(route_shuffled)
#                                 route_cap_after_shuffle = self.capacity_sum(route_shuffled)

#                                 # Update if you improved
#                                 if route_obj_after_shuffle<self.depot_close_time:
#                                     if route_cap_after_shuffle<self.vehicle_capacity:
#                                         if route_obj_after_shuffle < route_obj_before_shuffle:
# #                                             print(f"\n\route_shuffled {route_shuffled}")
# #                                             print(f"route {route}")
#                                             self.dictionary_harmony_memory[soln][route_idx]=route_shuffled
#                                             obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
#                                             self.fitness_dict[soln] = obj
#                                             counter_shuffle+=((route_obj_before_shuffle-route_obj_after_shuffle)/route_obj_before_shuffle)
    
                        # Insertion        
                        elif 0.3<=rand3<0.7:
                            # Choose two vehicles
                            veh1, veh2 = random.sample(self.dictionary_harmony_memory[soln].keys(), 2)
                            route1 = self.dictionary_harmony_memory[soln][veh1]
                            route2 = self.dictionary_harmony_memory[soln][veh2]

                            # Calculate time & capacity
                            route1_time_before_swap = self.calculate_current_time(route1)
                            route2_time_before_swap = self.calculate_current_time(route2)

                            # Obtain the position of the position of commited nodes
                            try:
                                limit1 = self.commited_nodes_dict[soln][veh1]

                            # Vehicle 1 was created with no commited nodes xD
                            except:
                                limit1 = 0

                            # Obtain the position of the position of commited nodes
                            try:
                                limit2 = self.commited_nodes_dict[soln][veh1]

                            # Vehicle 2 was created with no commited nodes xD
                            except:
                                limit2 = 0    


                            # Apply insertion_between
                            route1_inserted, route2_inserted = self.insertion_between(route1, route2, limit1, limit2)

                            # Calculate time & capacity
                            route1_time_after_swap = self.calculate_current_time(route1_inserted)
                            route2_time_after_swap = self.calculate_current_time(route2_inserted)
                            route1_cap_after_swap = self.capacity_sum(route1_inserted)
                            route2_cap_after_swap = self.capacity_sum(route2_inserted)

                            # if capacity is NOT violated
                            if route1_cap_after_swap <= self.vehicle_capacity and route2_cap_after_swap <= self.vehicle_capacity:
                            
                                if route1_time_after_swap<=self.depot_close_time and route2_time_after_swap<=self.depot_close_time:
                                    # if traveling time is BETTER than before
                                    if route1_time_after_swap+route2_time_after_swap < route1_time_before_swap+route2_time_before_swap:

                                        if len(route1_inserted) > 2:
                                            self.dictionary_harmony_memory[soln][veh1] = route1_inserted

                                        else:
                                            self.dictionary_harmony_memory[soln].pop(veh1)
                                            try:
                                                self.commited_nodes_dict[soln].pop(veh1)
                                            except:
                                                pass


                                        if len(route2_inserted) > 2:
                                            self.dictionary_harmony_memory[soln][veh2] = route2_inserted

                                        else: 
                                            self.dictionary_harmony_memory[soln].pop(veh2)

                                            try:
                                                self.commited_nodes_dict[soln].pop(veh2)
                                            except:
                                                pass

                                        # Update fitness dictionary
                                        obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
                                        self.fitness_dict[soln] = obj
                                        counter_insertion+=((route1_time_before_swap+route2_time_before_swap-route1_time_after_swap-route2_time_after_swap)/(route1_time_before_swap+route2_time_before_swap))
                                    
                        # Swap between routes      
                        else:
                            # Choose two vehicles
                            veh1, veh2 = random.sample(self.dictionary_harmony_memory[soln].keys(), 2)
                            route1 = self.dictionary_harmony_memory[soln][veh1]
                            route2 = self.dictionary_harmony_memory[soln][veh2]

                            # Calculate time & capacity
                            route1_time_before_swap = self.calculate_current_time(route1)
                            route2_time_before_swap = self.calculate_current_time(route2)

                            # Obtain the position of the position of commited nodes
                            try:
                                limit1 = self.commited_nodes_dict[soln][veh1]

                            # Vehicle 1 was created with no commited nodes xD
                            except:
                                limit1 = 0

                            # Obtain the position of the position of commited nodes
                            try:
                                limit2 = self.commited_nodes_dict[soln][veh1]

                            # Vehicle 2 was created with no commited nodes xD
                            except:
                                limit2 = 0    


                            # Apply insertion_between
                            route1_swapped, route2_swapped = self.swap_between(route1, route2, limit1, limit2)

                            # Calculate time & capacity
                            route1_time_after_swap = self.calculate_current_time(route1_swapped)
                            route2_time_after_swap = self.calculate_current_time(route2_swapped)
                            route1_cap_after_swap = self.capacity_sum(route1_swapped)
                            route2_cap_after_swap = self.capacity_sum(route2_swapped)

                            # if capacity is NOT violated
                            if route1_cap_after_swap <= self.vehicle_capacity and route2_cap_after_swap <= self.vehicle_capacity:
                                if route1_time_after_swap<=self.depot_close_time and route2_time_after_swap<=self.depot_close_time:
                                    # if traveling time is BETTER than before
                                    if route1_time_after_swap+route2_time_after_swap < route1_time_before_swap+route2_time_before_swap:

                                        if len(route1_swapped) > 2:
                                            self.dictionary_harmony_memory[soln][veh1] = route1_swapped

                                        else:
                                            self.dictionary_harmony_memory[soln].pop(veh1)
                                            try:
                                                self.commited_nodes_dict[soln].pop(veh1)
                                            except:
                                                pass


                                        if len(route2_swapped) > 2:
                                            self.dictionary_harmony_memory[soln][veh2] = route2_swapped

                                        else: 
                                            self.dictionary_harmony_memory[soln].pop(veh2)

                                            try:
                                                self.commited_nodes_dict[soln].pop(veh2)
                                            except:
                                                pass

                                        # Update fitness dictionary
                                        obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
                                        self.fitness_dict[soln] = obj   
                                        counter_swap_between+=((route1_time_before_swap+route2_time_before_swap-route1_time_after_swap-route2_time_after_swap)/(route1_time_before_swap+route2_time_before_swap))
                    ###########Else PAR enter VNS######################
                    
                    else:
#                     counter_not_enter_par+=1

                        # Find worst soln
                        worst_soln, worst_soln_obj = self.find_worst_soln()

                        # Use saved dictionary to get fitness
                        selected_route_obj = self.fitness_dict[soln]

                        # If randomly selected route is better than the worst, rewrite
                        if selected_route_obj < worst_soln_obj:
    #                             print(f"selected_route_obj {selected_route_obj} worst_soln_obj {worst_soln_obj}")
                            self.dictionary_harmony_memory[worst_soln]=copy.deepcopy(self.dictionary_harmony_memory[soln])
                            self.commited_nodes_dict[worst_soln]=copy.deepcopy(self.commited_nodes_dict[soln])
                            self.fitness_dict[worst_soln] = copy.deepcopy(self.fitness_dict[soln])                        

            
                #Else HMCR
                else:
                    counter_not_enter_par+=1

                    # Find worst soln
                    worst_soln, worst_soln_obj = self.find_worst_soln()

                    # Use saved dictionary to get fitness
                    selected_route_obj = self.fitness_dict[soln]

                    # If randomly selected route is better than the worst, rewrite
                    if selected_route_obj < worst_soln_obj:
#                             print(f"selected_route_obj {selected_route_obj} worst_soln_obj {worst_soln_obj}")
                        self.dictionary_harmony_memory[worst_soln]=copy.deepcopy(self.dictionary_harmony_memory[soln])
                        self.commited_nodes_dict[worst_soln]=copy.deepcopy(self.commited_nodes_dict[soln])
                        self.fitness_dict[worst_soln] = copy.deepcopy(self.fitness_dict[soln])
    ###########################VNS AT THE END OF EVERYTHING#####################
                #shaking operator: swap and insertion
                #lso: swap 10x

                condition_swap, condition_insertion = True, True
                route_idx = random.choice(list(self.dictionary_harmony_memory[soln]))
                route = self.dictionary_harmony_memory[soln][route_idx]

                while condition_swap==True: 

                    #iterate over neighborhood search with the first shaking operator 10 times
        #                             print('entered while')

#                     route = self.dictionary_harmony_memory[soln][route_idx]




                    if len(route)>4:

        #                                 print(f'entering again. for loop to run {self.neighbor_number} times')

                        route_obj_before_swap = self.calculate_current_time(route)



                        for i in range(self.neighbor_number):
        #######################here swap#######################################
                            # Swap
                            try:
                                x_prime = self.two_opt(route, self.commited_nodes_dict[soln][route_idx])

                            except:
                                x_prime = route

                            route_obj_after_swap = self.calculate_current_time(x_prime)
                            route_cap_after_swap =self.capacity_sum(x_prime)
        #####################end swap entering update mechanism###################################

                            # Update if you improved
                            if route_obj_after_swap < route_obj_before_swap:
                                if route_obj_after_swap<self.depot_close_time:
                                    if route_cap_after_swap<self.vehicle_capacity:                                           
                                        # if x prime better than x then repeat the process with the best solution and this stage (use current shaking operator)
                                        self.dictionary_harmony_memory[soln][route_idx]=x_prime
                                        obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
                                        self.fitness_dict[soln] = obj
                                        self.vns_2opt_counter+=1
                            else:
        #                                         print('hmmm 1')
                                condition_swap=False

                        condition_swap = False

                    else: 
        #                                 print('hmmm 2')
                        condition_swap=False

        #                         print('left while')                  

                while condition_insertion==True:  
        #                             route_idx = random.choice(list(self.dictionary_harmony_memory[soln]))


#                     route = self.dictionary_harmony_memory[soln][route_idx]
                    if len(route)>5:
                        route_obj_before_shuffle = self.calculate_current_time(route)
                        for i in range(self.neighbor_number):
                            try:
                                route_shuffled = self.shuffle_operator(route, self.commited_nodes_dict[soln][route_idx])

                            except KeyError:
                                route_shuffled = self.shuffle_operator(route, 0)

                            route_obj_after_shuffle = self.calculate_current_time(route_shuffled)
                            route_cap_after_shuffle = self.capacity_sum(route_shuffled)

                            # Update if you improved
                            if route_obj_after_shuffle < route_obj_before_shuffle:
                                if route_obj_after_shuffle<self.depot_close_time:
                                    if route_cap_after_shuffle<self.vehicle_capacity:


                                        self.dictionary_harmony_memory[soln][route_idx]=route_shuffled
                                        obj = sum([self.calculate_current_time(self.dictionary_harmony_memory[soln][x]) for x in self.dictionary_harmony_memory[soln].keys()])
                                        self.fitness_dict[soln] = obj
        #                                             counter_shuffle+=((route_obj_before_shuffle-route_obj_after_shuffle)/route_obj_before_shuffle)
                                        self.vns_insertion_counter+=1
                            else:
                                condition_insertion=False
                    else:
                        condition_insertion=False

    ############################ the end of vns########################################
        self.historic_improvement_operators['swap_between'].append(int(counter_swap_between*100))
        self.historic_improvement_operators['swap_within'].append(int(counter_swap_within*100))
        self.historic_improvement_operators['insertion_between'].append(int(counter_insertion*100))
        self.historic_improvement_operators['shuffle_within'].append(int(counter_shuffle*100))
#         print(f"vns was able to improve solution {self.counter_vns} times")
#         vns_improvement.append(self.counter_vns)
#     print(vns_improvement)
#         print(f"counter_swap_between {int(counter_swap_between*100)}%")
#         print(f"counter_swap_within {int(counter_swap_within*100)}%")
#         print(f"counter insertion {int(counter_insertion*100)}%")
#         print(f"counter shuffle {int(counter_shuffle*100)}%")
#         print(f"counter_not_enter_par {counter_not_enter_par}")
# data=dvrp(gen("c50d.csv", delimiter=','))


        print("vns_insertion_counter ", self.vns_insertion_counter)
        print("vns_2opt_counter ", self.vns_2opt_counter)
        print("\n\n")   
```
